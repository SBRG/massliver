```python
#import the library attributes
from bs4 import BeautifulSoup
import requests
import pandas as pd
from openpyxl import load_workbook, Workbook

url_list =[]
id_list = []
Kms = []
Vms = []
title=[]
protein_list = []
protein_list1=[]
full=[]
uni_id=[]
uni_id1=[]
```

```python
wb = load_workbook(filename = 'Sample Uniprot IDs.xlsx')
ws = wb['Sheet1']

for r in range(1, ws.max_row + 1):
    id = (ws.cell(row = r, column = 1).value)
    id_list.append(id)
id_list
```

```python
#name url 
for iD in id_list:
    url_list.append('https://www.uniprot.org/uniprot/' + iD)
# url_list
```

```python
#name url 
# ##My version

    
for url in url_list:


#make a GET request to get the raw HTML content
    html_content = requests.get(url).text

    #parse the html content to create a parse Tree object (soup)
    #soup --> BeautifulSoup object
    soup = BeautifulSoup(html_content, "lxml")

    #define ol tags 

    ol = soup.find_all('ol', class_='noNumbering kinetics') 
    for ol_ind in [0,1]: 
    #first ol tag (Kms)    
        if ol_ind == 0:
            for item in list(range(len(list(ol[ol_ind].children)))): 
    #define li tag
                li = list(ol[ol_ind].children)[item]
    #get value for Km
                Km = list(li.find("span",property = "text").children)[0]
                Kms.append(Km)
                protein = soup.find("h1",property = "name").get_text()
                protein_list.append(protein)
#                 uni_id.append(iD)
                u=url_list.index(url)
                uni= id_list[u]
                uni_id.append(uni)     

    #second ol tag (Vms)
        if ol_ind == 1:
            for item in list(range(len(list(ol[ol_ind].children)))): 
    #define li tag
                li = list(ol[ol_ind].children)[item]
    #get value for Vm
                Vm = list(li.find("span",property = "text").children)[0]
                Vms.append(Vm)       
                protein1 = soup.find("h1",property = "name").get_text()
                protein_list1.append(protein1)
                v=url_list.index(url)
                uni1= id_list[v]
                uni_id1.append(uni1)
## get PUBMED ID's

combined_Kms=[protein_list,Kms,uni_id]     
combined_Vms=[protein_list1,Vms,uni_id1]
```

```python
# ([Data[Kms].str.split(" " ,expand= True)])
Data_Kms= pd.DataFrame(combined_Kms).T

Data_Kms.columns= ["Enzyme","Combined","UniProt_ID's"]
Data_Kms[["Km","Unit","d","Substrate 1","Substrate 2"]]=Data_Kms["Combined"].str.split(" ",expand=True)
del Data_Kms['d']

Data_Kms
```

```python
Data_Vms= pd.DataFrame(combined_Vms).T

Data_Vms.columns= ["Enzyme","Combined","UniProt_ID's"]
Data_Vms[["Vm","Unit","d","e","Substrate1","Substrate2","i","h"]]=Data_Vms["Combined"].str.split(" ",expand=True)

del Data_Vms['d']
del Data_Vms['h']
del Data_Vms['i']
del Data_Vms['e']
##ADD UNIPROT ID's
Data_Vms
```
