```python
#import the library attributes
from bs4 import BeautifulSoup
import requests
import pandas as pd
from openpyxl import load_workbook, Workbook
# import gspread
# import df2gspread as d2g

url_list =[]
id_list = []
Kms = []
km=[]
Vms = []
title=[]
protein_list = []
full=[]
uni_id=[]
```

```python
wb = load_workbook(filename = 'Sample Uniprot IDs.xlsx')
ws = wb['Sheet1']

for r in range(1, ws.max_row + 1):
    id = (ws.cell(row = r, column = 1).value)
    id_list.append(id)

```

```python
#name url 
for iD in id_list:
    url_list.append('https://www.uniprot.org/uniprot/' + iD)
```

```python
# for url in url_list:

url='https://www.uniprot.org/uniprot/P30835'
#make a GET request to get the raw HTML content
html_content = requests.get(url).text

#parse the html content to create a parse Tree object (soup)
#soup --> BeautifulSoup object
soup = BeautifulSoup(html_content, "lxml")
```

```python

for url in url_list:
#make a GET request to get the raw HTML content
    html_content = requests.get(url).text
    
#parse the html content to create a parse Tree object (soup)
#soup --> BeautifulSoup object
    soup = BeautifulSoup(html_content, "lxml")
    
#define ol tags 
    ol = soup.find_all('ol', class_='noNumbering kinetics') 
    if ol == []:
        continue
     
#first ol tag (Kms)    
    for item in list(range(len(list(ol[0].children)))): 
#define li tag
        li = list(ol[0].children)[item]
#get value for Km
        Km = list(li.find("span",property = "text").children)[0]
        Kms.append(Km)
        k= "Km"
        Vms.append(k)
        protein = soup.find("h1",property = "name").get_text()
        protein_list.append(protein)
        u=url_list.index(url)
        uni= id_list[u]
        uni_id.append(uni)
    
#second ol tag (Vms)
        
    for item in list(range(len(list(ol[1].children)))): 
#define li tag
        li = list(ol[1].children)[item]
#get value for Vm
        Vm = list(li.find("span",property = "text").children)[0]
        Kms.append(Vm)       
            
        protein1 = soup.find("h1",property = "name").get_text()
        protein_list.append(protein1)
        v="Vm"
        Vms.append(v)
        v=url_list.index(url)
        uni1= id_list[v]
        uni_id.append(uni1)
        
        

combined_Kms=[protein_list,uni_id,Vms,Kms]     
```

```python
# ([Data[Kms].str.split(" " ,expand= True)])
Data_Kms= pd.DataFrame(combined_Kms).T

Data_Kms.columns= ["Enzyme","UniProt_ID's","Parameter","Combined",]
Data_Kms[["Combined","Substrate"]]=Data_Kms["Combined"].str.split("with",expand=True)
Data_Kms[["Data","Substrate","b"]]=Data_Kms["Combined"].str.split("for",expand=True)
Data_Kms[["Value","Unit","d","e"]]=Data_Kms["Data"].str.split(" ",expand=True)
del Data_Kms['e']

Data_Kms
```
